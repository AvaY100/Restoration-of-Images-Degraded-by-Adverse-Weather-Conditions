```python
import os
import cv2
import shutil
import numpy as np

# Define input and output directories
depth_dir = "/home/ubuntu/code/Marigold/output/data/input-depth-2/depth_bw"
color_dir = "/home/ubuntu/code/data/input"
gt_dir = "/home/ubuntu/code/data/gt"
output_dir = "/home/ubuntu/code/TransWeather/data/train"

# Create output directories if they don't exist
train_input_dir = os.path.join(output_dir, "input")
train_gt_dir = os.path.join(output_dir, "gt")
os.makedirs(train_input_dir, exist_ok=True)
os.makedirs(train_gt_dir, exist_ok=True)

# Iterate through files in depth directory
for depth_file in os.listdir(depth_dir):
    if depth_file.endswith((".png", ".jpg")):
        # Extract number from depth file name
        num = depth_file.split("_")[0]

        # Read depth image
        depth_img = cv2.imread(os.path.join(depth_dir, depth_file), cv2.IMREAD_UNCHANGED)

        # Read corresponding color image
        color_file = f"{num}_rain.png"
        color_img = cv2.imread(os.path.join(color_dir, color_file))

        # Check if color image exists
        if color_img is not None:
            # Resize depth image to match color image size
            depth_img_resized = cv2.resize(depth_img, (color_img.shape[1], color_img.shape[0]))

            # Map depth values to 0-255 range
            depth_img_mapped = (depth_img_resized / 65535.0) * 255.0
            depth_img_mapped = depth_img_mapped.astype(np.uint8)

            # Add depth as fourth channel to color image
            color_img_with_depth = np.dstack((color_img, depth_img_mapped))

            # Save the resulting image in the input directory
            output_file_input = os.path.join(train_input_dir, f"{num}_rain.png")
            cv2.imwrite(output_file_input, color_img_with_depth)

            # Copy corresponding gt image to gt directory
            shutil.copy(os.path.join(gt_dir, color_file), train_gt_dir)

            print(f"Processed: {depth_file}")

print("All files processed successfully.")
```

```python
import os
import cv2
import shutil
import numpy as np

# Define input and output directories
depth_dir = "/home/ubuntu/code/Marigold/output/data/input-depth-2/depth_bw"
color_dir = "/home/ubuntu/code/data/input"
gt_dir = "/home/ubuntu/code/data/gt"
output_dir = "/home/ubuntu/code/TransWeather/data/train"

# Create output directories if they don't exist
train_input_dir = os.path.join(output_dir, "input")
train_gt_dir = os.path.join(output_dir, "gt")
os.makedirs(train_input_dir, exist_ok=True)
os.makedirs(train_gt_dir, exist_ok=True)

def find_file_by_basename(directory, base_name):
    """Find a file in the specified directory that matches the base_name, ignoring file extension."""
    for file in os.listdir(directory):
        if os.path.splitext(file)[0] == base_name:
            return file
    return None  # Return None if no file matches

# Iterate through files in depth directory
for depth_file in os.listdir(depth_dir):
    if depth_file.endswith(".png"):
        # Extract base filename without the extension and remove '_pred' suffix
        base_name = os.path.splitext(depth_file)[0].replace('_pred', '')

        # Find corresponding color image
        color_file = find_file_by_basename(color_dir, base_name)
        if color_file:
            # Read depth image
            depth_img = cv2.imread(os.path.join(depth_dir, depth_file), cv2.IMREAD_UNCHANGED)

            # Read corresponding color image
            color_img = cv2.imread(os.path.join(color_dir, color_file))

            # Resize depth image to match color image size
            depth_img_resized = cv2.resize(depth_img, (color_img.shape[1], color_img.shape[0]))

            # Map depth values to 0-255 range
            depth_img_mapped = (depth_img_resized / 65535.0) * 255.0
            depth_img_mapped = depth_img_mapped.astype(np.uint8)

            # Add depth as fourth channel to color image
            color_img_with_depth = np.dstack((color_img, depth_img_mapped))

            # Save the resulting image in the input directory with .png extension
            output_file_input = os.path.join(train_input_dir, f"{base_name}.png")
            cv2.imwrite(output_file_input, color_img_with_depth)

            # Find and copy corresponding gt image to gt directory, ensuring output is .png
            gt_file = find_file_by_basename(gt_dir, base_name)
            if gt_file:
                shutil.copy(os.path.join(gt_dir, gt_file), os.path.join(train_gt_dir, f"{base_name}.png"))

            # print(f"Processed: {depth_file}")

print("All files processed successfully.")
```

```python
import os
import shutil
import random

# Define the directories for training and testing data
train_gt_dir = "/home/ubuntu/code/TransWeather/data/train/gt"
train_input_dir = "/home/ubuntu/code/TransWeather/data/train/input"
test_gt_dir = "/home/ubuntu/code/TransWeather/data/test/gt"
test_input_dir = "/home/ubuntu/code/TransWeather/data/test/input"

# Create test directories if they don't exist
os.makedirs(test_gt_dir, exist_ok=True)
os.makedirs(test_input_dir, exist_ok=True)

# Get list of files in the ground truth train directory
gt_files = os.listdir(train_gt_dir)

# Calculate 10% of the files
num_files_to_move = max(1, len(gt_files) // 10)

# Randomly select files to move
files_to_move = random.sample(gt_files, num_files_to_move)

# Move selected files from train to test directories
for file in files_to_move:
    # Move ground truth file
    shutil.move(os.path.join(train_gt_dir, file), os.path.join(test_gt_dir, file))
    
    # Move corresponding input file
    input_file = file  # Assuming the input file has the same name
    shutil.move(os.path.join(train_input_dir, input_file), os.path.join(test_input_dir, input_file))

print(f"Moved {num_files_to_move} files to test directories successfully.")
```
